```python
import pandas as pd
import matplotlib as mp
import numpy as np
```

```python
data = pd.read_csv('Sampled Data.csv')
```

```python
data.columns = data.columns.str.replace(' ', '_')
```

```python
data.head()
```

```python
# The following sets of codes displays the graph of the various parameters of the data.
# The salient feature of them is they almost remain constant throughout the year, with an accuracy of hour.
# These parameters can be used to define the reference keys for the experiment.
# The best among all is Surface Albedo, as it remains constant for a long interval of time
# The X-axis of the plots is the unique id, given to each data point, named as "Time."
# The convention used for naming can be used for any amount of data we have. It will remain unique.
# The unique key is time sorted.
```

```python
for i in range(12):
    for j in range(24):
        month = data[data.Month == i]
        hour = data[data.Hour == j]
        hour.Pressure.plot()
```

```python
mean = data.Pressure.mean()
variance = data.Pressure.var()
print('Mean - ', mean)
print('Variance - ', variance)
```
