```python
# Observations which can be derived from the above plots.
# Clearsky GHI remains almost constant in an hour in the month in January.
# Ozone concentration in the sky remains same through out the day but it varies on different days in the month.
# Solar Zenith Angle does not depend on the day, it depends only on the hour at which it is calculated. It is different for different hours in a day.
# Preciptible Water depends on the day and it remains constant in the day, at almost all hours.
# Dew Point also shows same trends as Preceptible Water.
# Surface Albedo remains constant for whole day and month.
```

```python
# DHI - Diffuse Horizontal Irradiance - The solar radiation that does not arrive on a direct path from the sun, but has scattered by the clouds and particles in the atmosphere and comes equally from all directions.
# DNI - Direct Normal Irradiance - The amount of light that is coming perpendicular to a surface. This type of irradiance belongs to rays that come in a striaght line from the direction of the sun at its current position in the sky.
# GHI - Global Horizontal Irradiance - Total amount of shortwave radiation recieved from above by a surface whihc is horizontal (parallel) to the ground. 
# GHI = DNI * cos(Solar Zenith Angle) + DHI
# Solar Zenith Angle is the angle between the sun's rays and the vertical direction.
# Albedo - is an expression of the ability of surfaces to reflect sunlight (heat from the sun)
```
